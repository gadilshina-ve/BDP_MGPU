```python
df = pd.read_csv("heart_attack_prediction_dataset.csv")

# printing first 6 raws of the dataframe
df.head(10)
```

```python
df.info()
```

```python
my_df = df[['Patient ID',
            'Age',
            'Sex',
            'Cholesterol',
            'Blood Pressure',
            'Heart Rate',
            'Diabetes',
            'Family History',
            'Smoking', 'Obesity',
            'Alcohol Consumption',
            'Exercise Hours Per Week',
            'Diet',
            'Previous Heart Problems',
            'Medication Use',
            'Stress Level',
            'BMI',
            'Country',
            'Continent',
            'Heart Attack Risk'
            ]].copy()
```

```python
df2 = my_df.dropna()
```

```python
df2.head(10)
```

```python
df2.isnull().values.any()
```

России нет в датасете, сделаем выборку по Японии.

```python
rslt_df = df2[df2['Country'] == "Japan"]
```

```python
rslt_df
```

```python
rslt_df.reset_index(drop=True, inplace=True)
```

```python
rslt_df
```

```python
rslt_df.info()
```

```python
rslt_df['Sex']=rslt_df['Sex'].str[:1]
```

```python
rslt_df.info()
```

```python
rslt_df
```

```python
rslt_df['Cholesterol'] = rslt_df['Cholesterol'].astype('float64')
```

```python
rslt_df.info()
```

```python
dd=rslt_df
```

```python
dd['Blood Pressure'] = [x.replace('/', ' / ') for x in dd['Blood Pressure']]
```

```python
dd.info()
```

```python
dd
```

```python
dd['BMI'] = dd['BMI'].astype('int64')
```

```python
dd.to_excel("heart_attack_prediction_dataset_Japan.xlsx", sheet_name='heart_attack_prediction_dataset_Japan')
from google.colab import files
files.download('heart_attack_prediction_dataset_Japan.xlsx')
```
