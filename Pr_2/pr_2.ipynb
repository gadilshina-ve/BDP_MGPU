Москва на каком месте?

```python
rslt_df = df2[df2['City'] == "Moscow"]
```

```python
rslt_df
```

```python
rslt_df.reset_index(drop=True, inplace=True)
```

```python
rslt_df
```

```python
rslt_df.info()
```

```python
rslt_df['City']=rslt_df['City'].str[:3]
```

```python
rslt_df.info()
```

```python
rslt_df
```

```python
rslt_df['Outdoor activities(City)'] = rslt_df['Outdoor activities(City)'].astype('float64')
```

```python
rslt_df.info()
```

```python
dd=rslt_df
```

```python
dd['Cost of a bottle of water(City)'] = [x.replace('£', '$') for x in dd['Cost of a bottle of water(City)']]
```

```python
dd.info()
```

```python
dd
```

```python
dd['Number of take out places(City)'] = dd['Number of take out places(City)'].astype('float64')
```

```python
dd.to_excel("healthy_lifestyle_city_2021_Moscow.xlsx", sheet_name='healthy_lifestyle_city_2021_Moscow')
from google.colab import files
files.download('healthy_lifestyle_city_2021_Moscow.xlsx')
```
